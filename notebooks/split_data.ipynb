```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../../datasets/jigsaw-multilingual-toxic-comment-classification/jigsaw-toxic-comment-train.csv')
df.head()
```

```python
train, test = train_test_split(df[['id', 'comment_text', 'toxic']], test_size=0.05)
val, test = train_test_split(test, test_size=0.5)
train.to_csv('train_data.csv', index=False)
val.to_csv('val_data.csv', index=False)
test.to_csv('test_data.csv', index=False)
```

```python
def count_data(df):
    print('0:', len(df[df['toxic'] == 0]), '1:', len(df[df['toxic'] == 1]))
```

```python
count_data(train)
count_data(test)
count_data(val)
```

```python
def count_data(df):
    print('0:', len(df[df['toxic'] < 0.5]), '1:', len(df[df['toxic'] >= 0.5]))
```

```python
df2 = pd.read_csv('../../datasets/jigsaw-multilingual-toxic-comment-classification/jigsaw-unintended-bias-train.csv')
df2.head()
```
